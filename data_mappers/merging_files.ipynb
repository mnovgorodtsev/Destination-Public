```python
pollution_path  = 'data/env_air_emis_page_spreadsheet.xlsx'
pollution_df = pd.read_excel(pollution_path, sheet_name=3)
pollution_df
```

```python
pollution_df = pollution_df[['TIME', '2021']]
pollution_df
```

```python
import pycountry

def get_country_code(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_2
    except LookupError:
        return None  

pollution_df['CNTR_CODE'] = pollution_df['TIME'].apply(get_country_code)

print(pollution_df)
```

```python
merged_df = pd.read_csv('merged_crimes.csv')
merged_df = pd.merge(merged_df, pollution_df, on='CNTR_CODE', how='left')
print(merged_df)
```

```python
merged_df.to_csv('merged_crimes_and_poll.csv')
```
